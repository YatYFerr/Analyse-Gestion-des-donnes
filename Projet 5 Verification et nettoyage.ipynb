## Vérification et nettoyage des données - Projet 5

```python
import pandas as pd
import numpy as np
import os 
import matplotlib.pyplot as plt
%matplotlib inline 

folder = "/Users/yathyarayamauchiferreira/Documents/OpenClassrooms/Projet 5/"
erp = folder + "erp.xlsx"
web = folder + "web.xlsx"
liaison = folder + "liaison.xlsx"
```

### Traitement du Fichier WEB

#### Importation et lecture du fichier EXCEL

```python
df_web = pd.read_excel(web)
df_web.head()
```

#### Vérification du nombre de lignes

```python
df_web.shape
```

#### Vérification de type de colonne et des valeurs manquantes

```python
df_web.info()
```

#### Droper les colonnes pas utiles sur le df web:

```python
df_web.drop(columns=['virtual', 'downloadable','rating_count','average_rating','tax_status','tax_class',
                     'post_content_filtered','post_mime_type','comment_count','post_content','post_password'],inplace = True)
df_web.head()
```

#### Remplacer le nom de la colonne sku par product_id pour hegemonization des df:

```python
df_web.rename(columns={"sku": "id_web"},inplace = True)
df_web.head()
```

#### Faire un tri sur les lignes qu'on veut garder:

```python
df_web.sort_values(by = ['id_web','post_type'],ascending = False)
df_web.head()
```

#### Effacer les lignes en double:

```python
df_web.drop_duplicates(subset=['id_web'],inplace = True)
df_web.head()
```

#### Verifier le nombre de lignes:

```python
df_web.shape
```

#### Vérification de type de colonne et des valeurs manquantes:

```python
df_web.head()
```

#### Droper les valeurs nulls sur la colonne Id_product:

```python
df_web.dropna(subset=['id_web'],inplace = True)
df_web.head()
```

```python
df_web.info()
```

### Importation et lecture du fichier ERP

```python
df_erp = pd.read_excel(erp)
df_erp.head()
```

#### Vérification du nombre de lignes:

```python
df_erp.shape
```

#### Vérification de type de colonne et des valeurs manquantes:

```python
df_erp.info()
```

### Importation et lecture du fichier LIAISON

```python
df_liaison = pd.read_excel(liaison)
df_liaison.head()
```

```python
df_liaison.shape
```

```python
df_liaison.info()
```

```python
df_liaison.dropna(inplace = True)
df_liaison.info()
```

#### Jointure interne (left merge) des deux df ERP et Liaison avec la clé product_id:

```python
df_erp_lia = pd.merge(df_erp, df_liaison, how = 'left', on=["product_id"])
df_erp_lia.head()
```

```python
df_erp_lia.info()
```

**Produits qui ne sont pas sur le web:**

```python
df_product_not_in_web = df_erp_lia[df_erp_lia['id_web'].isna()]
df_product_not_in_web.info()
```

**Traitement des valeurs vides:**

```python
df_erp_lia.dropna(inplace=True)
df_erp_lia.info()
```

#### Jointure (left merge) du nouveau df erp_lia avec le df web:

```python
df_final = pd.merge(df_erp_lia, df_web, how='left',on=['id_web'])
df_final.head()
```

```python
df_final.info()
```

```python
df_final[df_final['total_sales'].isna()]
```

**Ces produits existent sur la partie web mais par sur l'ERP**

**Droper les valeurs nulls:**

```python
df_final.dropna(how = 'any',inplace = True)
df_final.info()
```

#### générer le fichier Df_final excel traité

```python
folder = "/Users/yathyarayamauchiferreira/Documents/OpenClassrooms/Projet 5/"
df_final.to_excel(folder + 'df_final.xlsx', index=False)
```
